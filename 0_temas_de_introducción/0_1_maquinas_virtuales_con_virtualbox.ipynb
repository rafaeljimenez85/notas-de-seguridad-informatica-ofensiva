#### Configurar una red en NAT

Ya tenemos virtualbox y pack de extensiones instalado. Con el fin de seguir mejor los ejemplos de este manual, vamos a configurar una red en [NAT](https://es.wikipedia.org/wiki/Traducci%C3%B3n_de_direcciones_de_red). Para ello, seleccionamos la opción *Preferences...* del menú *File*:

<img src="img/virtualbox_15.png" style="float:left">
<div style="clear:both"></div><br>

En el menú de opciones, seleccionamos *Network* en el menú de la izquierda y pulsamos el botón que tiene un signo **+**:

<img src="img/virtualbox_16.png" style="float:left">
<div style="clear:both"></div><br>

Se creará un nuevo *NatNetwork*. Ahora pulsamos en el botón con el icono engranaje para configurarlo:

<img src="img/virtualbox_17.png" style="float:left">
<div style="clear:both"></div><br>

Establecemos la red que queremos -para seguir los ejemplos del libro se ha elegido la red *192.168.119.0/24*- y pulsamos el botón *OK*:

<img src="img/virtualbox_18.png" style="float:left">
<div style="clear:both"></div><br>

#### Configurando el servicio snmp

Iniciamos la máquina *Metasploitable* pulsando el botón *Start*:

<img src="img/metasploitable2_13.png" style="float:left">
<div style="clear:both"></div><br>

Cuando apareca la pantalla de logín, vemos que nos indica que las credenciales son:

  * **Usuario:** msfadmin
  * **Contraseña:** msfadmin
  
<img src="img/metasploitable2_14.png" style="float:left">
<div style="clear:both"></div><br>

Nos logamos en la máquina usando esas credenciales:

<img src="img/metasploitable2_15.png" style="float:left">
<div style="clear:both"></div><br>

Ahora editaremos la configuración del servicio snmp para que, en vez de sólo estar a la escucha en localhost, lo este en la red NAT que hemos creado. Para editar la configuración del servicio snmp, editamos el fichero */etc/default/snmpd* con cualquier editor de texto. En este caso usamos [nano](https://www.nano-editor.org/):

<img src="img/metasploitable2_16.png" style="float:left">
<div style="clear:both"></div><br>

Autorizamos la operación con la contraseña *msfadmin*:

<img src="img/metasploitable2_17.png" style="float:left">
<div style="clear:both"></div><br>

Una vez se abrá el fichero, tenemos que localizar la dirección IP del localhost (127.0.0.1) de la línea que empieza por *SNMPDOPTS*:

<img src="img/metasploitable2_18.png" style="float:left">
<div style="clear:both"></div><br>

Una vez localizada, borramos la dirección IP. La línea debería quedar así:

<img src="img/metasploitable2_19.png" style="float:left">
<div style="clear:both"></div><br>

Una vez modificada la línea pulsamos la combinación de teclas *Ctrl+X* para salir:

<img src="img/metasploitable2_20.png" style="float:left">
<div style="clear:both"></div><br>

Pulsamos la tecla **Y** para confirmar que queremos guardar los cambios que hemos realizado al fichero:

<img src="img/metasploitable2_21.png" style="float:left">
<div style="clear:both"></div><br>

Finalmente pulsamos la tecla *Intro* para confirmar y salir:

<img src="img/metasploitable2_22.png" style="float:left">
<div style="clear:both"></div><br>

Una vez realizado los cambios en el fichero podemos iniciar el servicio snmp tecleando: *sudo /etc/init.d/snmpd start* y autorizando la operación con nuestra contraseña (*msfadmin*):

<img src="img/metasploitable2_23.png" style="float:left">
<div style="clear:both"></div><br>

El servicio quedará iniciado:

<img src="img/metasploitable2_24.png" style="float:left">
<div style="clear:both"></div><br>

Necesitaremos iniciar este servicio en la fase de enumeración. Para detener el servicio snmp tecleamos: *sudo /etc/init.d/snmpd stop*:

<img src="img/metasploitable2_25.png" style="float:left">
<div style="clear:both"></div><br>

Cuando terminemos de usar la máquina elegimos la opción *Close...* del menú *File*:

<img src="img/metasploitable2_26.png" style="float:left">
<div style="clear:both"></div><br>

Seleccionamos la opción *Power off the machine* y pulsamos el botón *OK*:

<img src="img/metasploitable2_27.png" style="float:left">

#### Comprobando el hash de la descarga

Sabemos que cuando descargamos un archivo desde internet pasa por varios ordenadores antes de llegar al nuestro, para comprobar la integridad de la descarga. Es decir, para verificar que el fichero que nos hemos descargado no ha sufrido modificaciones mientras pasaba por los ordenadores intermedios y asegurarnos que tenemos exactamente el mismo fichero que hay en la web ubuntu, vamos a calcular el hash del fichero descargado. Para ello, primero necesitamos saber cual es el hash del fichero que esta en la web de ubuntu. Para conocer el hash del fichero original, accedemos a la [web de descargas](http://releases.ubuntu.com/) de ubuntu y hacemos click sobre la versión que nos hayamos descargado. En este caso Ubuntu *16.04.3 LTS*:

<img src="img/ubuntu_server_3.png" style="float:left">
<div style="clear:both"></div><br>

Hacehmos click sobre el algoritmo hash que queramos usar para la verificación. Se recomienda usar *SHA256SUMS* porque tanto para [SHA1](https://security.googleblog.com/2017/02/announcing-first-sha1-collision.html?m=1) como para MD5 se han encontrado colisiones y no son algoritmos de hashing seguros.

<img src="img/ubuntu_server_4.png" style="float:left">
<div style="clear:both"></div><br>

Tras hacer click podremos ver el hash del fichero que esta en el servidor. El que nos interesa es el de la versión server de ubuntu que hemos descargado:

<img src="img/ubuntu_server_5.png" style="float:left">
<div style="clear:both"></div><br>

Ahora comprobamos que el hash que tiene el fichero descargado. Para calcular el hash del fichero descargado podemos usar [powershell](https://es.wikipedia.org/wiki/Windows_PowerShell). Desde el interprete de comandos de windows, tecleamos *powershell*:

<img src="img/ubuntu_server_6.png" style="float:left">
<div style="clear:both"></div><br>

Veremos que estamos usando *powershell* porque al comienzo del [promtp](http://infohost.nmt.edu/tcc/help/unix/shellprompt.html) del interprete de comandos, aparecen las letras *PS*:

Podemos usar el comando ([cmdlet](https://technet.microsoft.com/en-us/library/dd772285.aspx)) Get-FileHash. Una forma de usarlo es indicarle fichero sobre el que queremos calcular el hash y algoritmo a utilizar. En este caso usaremos SHA256. Desde la carpeta donde hayamos descargado el archivo, ejecutamos el comando:

*Get-FileHash '.\ubuntu-16.04.3-server-amd64.iso' -Algorithm SHA256*

<img src="img/ubuntu_server_7.png" style="float:left">
<div style="clear:both"></div><br>

Si queremos ver más opciones sobre el cmdlet, [Get-FileHash](https://docs.microsoft.com/en-us/powershell/module/microsoft.powershell.utility/get-filehash?view=powershell-5.1), podemos usar el comando (cmdlet) Get-Help para obtener una breve información:

<img src="img/ubuntu_server_8.png" style="float:left">
<div style="clear:both"></div><br>

Una vez obtenidos los 2 hashes, tenemos que compararlos para saber que el fichero en origen es el mismo que tenemos descargado. Para ello creamos una variable en *powershell* que contenga el hash del fichero original. Como las letras de este hash están en minúsculas y Get-FileHash nos lo ha calculado en mayúsculas. Al guardar el valor en la variable, lo convertimos en mayúsculas usando el método ToUpper():

<img src="img/ubuntu_server_9.png" style="float:left">
<div style="clear:both"></div><br>

Ahora creamos otra variable para guardar el valor hash del fichero descargado que se ha calculado desde *powershell* usando el comando (cmdlet) Get-FileHash:

<img src="img/ubuntu_server_10.png" style="float:left">
<div style="clear:both"></div><br>

Ahora comparamos ambos hashes y vemos que son iguales:

<img src="img/ubuntu_server_11.png" style="float:left">
<div style="clear:both"></div><br>

Con esto confirmamos que el fichero no ha sido modificado durante el transcurso de la descarga. Es decir, que el fichero se ha descargado correctamente.
